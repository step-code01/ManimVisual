```python
%%writefile function_visual.py
from manim import *
import numpy as np

def modulusfn(x):
  if x>=0:
    return x
  else:
    return -x

class FunctionVisual(Scene):
  def construct(self):
    #Drawing functions:
      axes = Axes(x_range=[-10,10],y_range=[-3,10],axis_config={"include_tip": False})
      modulus_curve = axes.plot(modulusfn,color=RED_B)
      modulus_label = axes.get_graph_label(modulus_curve,label = MathTex(r"\left|x\right|"),x_val=4) # |x| and mod(x) causes render issues since not wrapped properly in LaTeX (rawstring)
      #x_val means...to position the label visually where you want it on the curve.

      self.play(Create (axes),Create(modulus_curve),Write(modulus_label))
      self.wait(2)
```

```python
%%writefile power_series.py
from manim import *
import numpy as np

class PowerSeriesIntuition(Scene):
    def construct(self):
        # Title
        title = Text("What is a Power Series?", font_size=48)
        self.play(Write(title))
        self.wait(2)
        self.play(FadeOut(title))

        # Draw exponential function
        axes = Axes(
            x_range=[-3, 3],
            y_range=[0, 20],
            axis_config={"include_tip": False},
        )
        exp_curve = axes.plot(lambda x: np.exp(x), color=BLUE)
        exp_label = axes.get_graph_label(exp_curve, label=MathTex("e^x"))

        self.play(Create(axes), Create(exp_curve), Write(exp_label))
        self.wait(2)

        # Tangent line at x = 0
        tangent = axes.plot(lambda x: 1 + x, color=YELLOW)
        tangent_label = axes.get_graph_label(tangent, label=MathTex("1 + x"), x_val=1, direction=UP)

        self.play(Create(tangent), Write(tangent_label))
        self.wait(2)

        # Quadratic term
        quadratic = axes.plot(lambda x: 1 + x + x**2 / 2, color=GREEN)
        quadratic_label = axes.get_graph_label(quadratic, label=MathTex("1 + x + \\frac{x^2}{2}"))

        self.play(Transform(tangent, quadratic), Transform(tangent_label, quadratic_label))
        self.wait(2)

        # Cubic term
        cubic = axes.plot(lambda x: 1 + x + x**2 / 2 + x**3 / 6, color=ORANGE)
        cubic_label = axes.get_graph_label(cubic, label=MathTex("1 + x + \\frac{x^2}{2} + \\frac{x^3}{6}"))

        self.play(Transform(tangent, cubic), Transform(tangent_label, cubic_label))
        self.wait(2)

        # Full power series text
        series_text = MathTex("f(x) = \\sum_{n=0}^{\\infty} \\frac{x^n}{n!}")
        self.play(FadeOut(axes), FadeOut(tangent), FadeOut(tangent_label), FadeOut(exp_curve), FadeOut(exp_label))
        self.play(Write(series_text))
        self.wait(3)

        # End
        self.play(FadeOut(series_text))
```
